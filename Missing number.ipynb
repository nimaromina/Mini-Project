```python
lst = [2,0,3,1,6,4]
def find_missing_num(lst):
    n = len(lst)+1
    _sum = n * (n-1) // 2
    return _sum-sum(lst)
```

```python
find_missing_num(lst)
```

```python
lst = [2,0,3,1,6,4]
def find_missing_num2(lst):
    lst.sort()
    i = 0
    for num in lst:
        if i != num:
            print(i)
            break
        i += 1
```

```python
find_missing_num2(lst)
```

```python
%timeit find_missing_num(lst)
```

```python
%timeit find_missing_num2(lst)
```
